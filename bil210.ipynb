 # ONE HOT ENCODING

Get_dummies kullanarak renk değişkeninimizin özniteliklerini bulalım. Amacımız renk değişkenimizi nominalden sayısal değer çevirmek.

```python
df["Renk"].unique()
```

```python
df['Renk'] = pd.Categorical(df['Renk'])
dfDummies = pd.get_dummies(df['Renk'], prefix = 'Renk')
dfDummies
```

Komşu sayısının 2'den 15'e kadar olan her sayıyla komşu sayısı olarak deneyerek her denemede skoru kaydetmemiz istenmiştir.

```python
knn_params = {"n_neighbors": np.arange(2,15)}
```

```python
knn = KNeighborsClassifier()
knn_cv = GridSearchCV(knn, knn_params, cv = 3)
knn_cv.fit(x_train, y_train)
```

Komşu sayısının miktarının değişmesiyle skorunun da değişeceği öngörülmüştü, bunu görüntüleyebilmek için modelin skorunun plot olarak gösterelim.

```python
score_list = []

for each in range(2,15):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test, y_test))
    
plt.plot(range(2,15),score_list)
plt.xlabel("k değerleri")
plt.ylabel("doğruluk skoru")
plt.show()
```

```python
print("En iyi skor: " + str(knn_cv.best_score_))
print("En iyi parametreler: " + str(knn_cv.best_params_))
```
